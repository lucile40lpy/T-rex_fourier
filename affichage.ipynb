```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def create_segment(p1, p2, m):# p1 et p2 pas lise mais complexe
    div = (p2-p1)/m
    points = [0]*(m+1)
    for i in range(m+1):
        points[i]=p1 + i*div
    return points

points = create_segment(complex(2,3),complex(10,4),5)
z = complex(1, 2)
print(z.real)
```

```python
def show_figure(liste_points, m):

    for i in range(len(liste_points) - 1):
        x = []
        y = []
        seg = create_segment(liste_points[i], liste_points[i+1], m)
    
        liste_x = []
        liste_y = []
        for p in seg:
            liste_x.append(p.real)
            liste_y.append(p.imag)

    
        x = np.asarray(liste_x)
        y = np.asarray(liste_y)

        plt.scatter(x, y)

    plt.show()

l = [complex(0, 0), complex(1, 0), complex(1,1), complex(0, 1), complex(0,0)]
show_figure(l, 100)
```
