```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def digit_shape(d):
    if d==0:
        return [1,0,1j,1j+1]
    
def segment(p1,p2,m):
    div = (p2-p1)/m
    points = [0]*(m+1)
    for i in range(m+1):
        points[i] = p1 + i*div
    return points

segment(0,10,5)

def joli(d, m):
    coord = digit_shape(d)
    for (x,y) in coord:
        seg = segment(x, y, m)

```
