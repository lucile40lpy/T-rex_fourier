```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def digit_shape(d):
    if d==0:
        return [1,0,1j,1j+1]
    
def segment(p1,p2,m):
    div = (p2-p1)/m
    points = [0]*(m+1)
    for i in range(m+1):
        points[i] = p1 + i*div
    return points

segment(0,10,5)


```

```python
def joli(d, m):
    coord = digit_shape(d)
    for i in len(coord):
        seg = segment(coord[i], coord[i+1], m)
        seg = np.array(seg)
        fig,ax = plt.subplots()
        ax.scatter(seg.real,seg.imag)
    print(fig)
```

```python
joli(0, 100)
```
