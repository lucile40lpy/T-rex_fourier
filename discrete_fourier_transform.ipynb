```python
import math
```

```python
def dft(complexes, n):
    
    coef_fourier = []
    N = len(complexes)
    
    # On boucle sur le nombre de complexes
    for i in range(n):
        for k in range(N):
            c_i = ( complexes[k] * math.exp( -(2 * i * math.pi * n * k) / N ) )
        
        coef_fourier.append( c_i/N )
        
    return coef_fourier
```
